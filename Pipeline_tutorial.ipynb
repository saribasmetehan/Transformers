```python
from transformers import pipeline
```

```python
import pandas as pd
```

**Sentiment Analysis**

```python
sent_classifier = pipeline("sentiment-analysis")
```

```python
text = "Mustafa Kemal Atatürk, the founder of modern Turkey, is celebrated for his visionary leadership and extensive reforms that transformed the country into a secular, progressive nation."
```

```python
output = sent_classifier(text)
pd.DataFrame(output)
```

**Zero-Shot Classification**

```python
zero_classifier = pipeline("zero-shot-classification")
```

```python
labels = ["tech","business","history"]
```

```python
output = zero_classifier(text,labels)
pd.DataFrame(output)
```

**Named Entity Recognition**

```python
ner = pipeline("ner", grouped_entities = True)
```

```python
text_ner = "Elon Reeve Musk was born on June 28, 1971, in Pretoria, South Africa's administrative capital. He is of British and Pennsylvania Dutch ancestry. His mother, Maye Musk (née Haldeman), is a model and dietitian born in Saskatchewan, Canada, and raised in South Africa."
```

```python
output_ner = ner(text_ner)
pd.DataFrame(output_ner)
```

**Question Answering**

```python
reader = pipeline("question-answering")
```

```python
text_qa = "Mustafa Kemal Atatürk, the founder of modern Turkey, is celebrated for his visionary leadership and extensive reforms that transformed the country into a secular, progressive nation."
question = "Who is Atatürk"
```

```python
output_qa = reader(question=question, context = text_qa)
pd.DataFrame([output_qa])
```

**Translate**

```python
translator = pipeline("translation_en_to_de")
```

```python
text_tr = "Mustafa Kemal Atatürk, the founder of modern Turkey, is celebrated for his visionary leadership and extensive reforms that transformed the country into a secular, progressive nation."
```

```python
output_tr = translator(text_tr)
```

```python
print(output_tr[0]["translation_text"])
```

```python
translator_hel = pipeline("translation", model = "Helsinki-NLP/opus-mt-tc-big-en-tr")
```

```python
output_hel = translator_hel(text_tr)
```

```python
print(output_hel)
```
